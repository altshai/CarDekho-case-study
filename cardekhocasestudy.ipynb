Which is the most and least cost depreciated vehicle in data ?


```python
# Calculate the depreciation for each vehicle
df['Depreciation'] = df['Present_Price'] - df['Selling_Price']

# Find the most depreciated vehicle
most_depreciated_vehicle = df.loc[df['Depreciation'].idxmax()]
print(f"The most depreciated vehicle is {most_depreciated_vehicle['Car_Name']} with a depreciation of {most_depreciated_vehicle['Depreciation']}.")

# Find the least depreciated vehicle
least_depreciated_vehicle = df.loc[df['Depreciation'].idxmin()]
print(f"The least depreciated vehicle is {least_depreciated_vehicle['Car_Name']} with a depreciation of {least_depreciated_vehicle['Depreciation']}.")
```

The most depreciated vehicle is land cruiser with a depreciation of 57.599999999999994.
The least depreciated vehicle is Honda Activa 4G with a depreciation of 0.030000000000000027.

___

Which brands of vehicles are less affected by cost depreciation ?


```python
# Calculate the average depreciation for each brand
average_depreciation = df.groupby('Car_Name')['Depreciation'].mean()

# Sort the brands by average depreciation in ascending order
less_affected_brands = average_depreciation.sort_values()

# Display the brands that are less affected by cost depreciation
print("Brands of vehicles that are less affected by cost depreciation:")
print(less_affected_brands.head(10))  # Display the top 10 brands with the least depreciation

# Plotting the brands with the least depreciation
less_affected_brands.head(10).plot(kind='barh', figsize=(10, 5))
plt.xlabel('Average Depreciation')
plt.ylabel('Car Name')
plt.title('Top 10 Brands of Vehicles Less Affected by Cost Depreciation')
plt.show()
```

Are there any factors which you feel affect the cost depreciation ?


```python
# Exclude non-numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Display the correlation of Depreciation with other factors
depreciation_correlation = correlation_matrix['Depreciation'].sort_values(ascending=False)
print("Correlation of Depreciation with other factors:")
print(depreciation_correlation)

# Plotting the correlation of Depreciation with other factors
depreciation_correlation.plot(kind='bar', figsize=(10, 5))
plt.xlabel('Factors')
plt.ylabel('Correlation with Depreciation')
plt.title('Correlation of Depreciation with Other Factors')
plt.show()
```

In general selling price is affected by age of vehicle and distance driven by vehicle , is it observable from data ?


```python
# Calculate the age of the vehicle
df['Vehicle_Age'] = 2023 - df['Year']

# Scatter plot of Selling Price vs. Vehicle Age
plt.figure(figsize=(10, 5))
plt.scatter(df['Vehicle_Age'], df['Selling_Price'], alpha=0.5)
plt.xlabel('Vehicle Age (years)')
plt.ylabel('Selling Price')
plt.title('Selling Price vs. Vehicle Age')
plt.show()

# Scatter plot of Selling Price vs. Kms Driven
plt.figure(figsize=(10, 5))
plt.scatter(df['Kms_Driven'], df['Selling_Price'], alpha=0.5)
plt.xlabel('Kms Driven')
plt.ylabel('Selling Price')
plt.title('Selling Price vs. Kms Driven')
plt.show()

# Calculate correlation coefficients
correlation_age = df['Selling_Price'].corr(df['Vehicle_Age'])
correlation_kms = df['Selling_Price'].corr(df['Kms_Driven'])
print(f"Correlation between Selling Price and Vehicle Age: {correlation_age}")
print(f"Correlation between Selling Price and Kms Driven: {correlation_kms}")
```

Do you find any deal in two wheelers which exceeded the general expectation ? Can you find reason for it ?


```python
# Calculate the ratio of Selling Price to Present Price for two-wheelers
two_wheelers_df['Price_Ratio'] = two_wheelers_df['Selling_Price'] / two_wheelers_df['Present_Price']

# Identify deals that exceeded general expectation (Price_Ratio significantly different from 1)
unexpected_deals = two_wheelers_df[(two_wheelers_df['Price_Ratio'] > 1.2) | (two_wheelers_df['Price_Ratio'] < 0.8)]

# Display the unexpected deals
print("Deals in two-wheelers which exceeded general expectation:")
print(unexpected_deals)

# Analyze the reasons for these deals
for index, deal in unexpected_deals.iterrows():
    print(f"\nDeal Analysis for {deal['Car_Name']}:")
    print(f"Year: {deal['Year']}")
    print(f"Selling Price: {deal['Selling_Price']}")
    print(f"Present Price: {deal['Present_Price']}")
    print(f"Kms Driven: {deal['Kms_Driven']}")
    print(f"Fuel Type: {deal['Fuel_Type']}")
    print(f"Seller Type: {deal['Seller_Type']}")
    print(f"Transmission: {deal['Transmission']}")
    print(f"Owner: {deal['Owner']}")
    print(f"Depreciation: {deal['Depreciation']}")
    print(f"Vehicle Age: {deal['Vehicle_Age']}")
    print(f"Price Ratio: {deal['Price_Ratio']}")

# Plotting the unexpected deals
plt.figure(figsize=(10, 5))
plt.scatter(two_wheelers_df['Present_Price'], two_wheelers_df['Selling_Price'], alpha=0.5, label='All Deals')
plt.scatter(unexpected_deals['Present_Price'], unexpected_deals['Selling_Price'], color='red', label='Unexpected Deals')
plt.xlabel('Present Price')
plt.ylabel('Selling Price')
plt.title('Unexpected Deals in Two-Wheelers')
plt.legend()
plt.show()
```
